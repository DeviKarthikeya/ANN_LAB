```python
import numpy as np
import matplotlib.pyplot as plt
def linear(x):
  return x;
def sigmoid(x):
  return 1/(1+np.exp(-x))
def tanh(x):
  return np.tanh(x)
def relu(x):
  return np.maximum(0,x)
def leaky_relu(x):
  return np.maximum(0.01*x,x)
def biploar(x):
  return (np.exp(x)-np.exp(-x))/(np.exp(x)+np.exp(-x))
def softmax(x):
  return np.exp(x)/np.sum(np.exp(x))

x = np.linspace(-10, 10, 50)
plt.figure(figsize=(10,10))
plt.subplot(2, 2, 1)
plt.plot(x,linear(x))
plt.title("Linear")
plt.show()
plt.plot(x,sigmoid(x))
plt.title("Sigmoid")
plt.show()
plt.plot(x,tanh(x))
plt.title("tanh")
plt.show()
plt.plot(x,relu(x))
plt.title("relu")
plt.show()
plt.plot(x,leaky_relu(x))
plt.title("leaky_relu")
plt.show()
plt.plot(x,biploar(x))
plt.title("biploar sigmoid")
plt.show()
plt.plot(x,softmax(x))
plt.title("Softmax")
plt.show()
```
